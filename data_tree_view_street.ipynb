```python
#nos imports
import pandas as pd
import os
import urllib
```

```python
#lecture de la data
input_filename = 'new_arbres.csv'
data = pd.read_csv(input_filename, encoding='utf-8',sep = ',') 
data.head(150)
```

```python
#on plot par essence et coordonnées respectives
import matplotlib.pyplot as plt
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 6)

df_tree = data[data.latitude > 37.6]

# Look at secondary caretakers, looking specifically for FuF
df_tree.groupby('LIBELLEFRANCAIS').plot(kind='scatter', x='longitude', y='latitude')
```

```python
#sur la ville de paris
import seaborn as sns
sns.set(style="ticks")

# Pretty plot with seaborn to look at tree distributions
sns.jointplot(df_tree.longitude, df_tree.latitude, kind="hex", color="#4CB391", size=20)
```
