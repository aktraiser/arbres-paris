```python
#nos imports
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
import seaborn as sns
```

```python
#lecture de la nouvelle data
input_filename = 'new_arbres.csv'
df = pd.read_csv(input_filename, encoding='utf-8',sep = ',') 
df.head(150)
```

```python
df.LIBELLEFRANCAIS.value_counts().index.tolist()
```

```python
df.LIBELLEFRANCAIS.value_counts()
```

```python
#on cherche a définir la proportion en arbre coupés et non coupés
def is_age(x):
    return x > 0

df['age'] = df['CIRCONFERENCEENCM'].apply(is_age)
```

```python
df['age'] = df['CIRCONFERENCEENCM'].apply(lambda x: x > 0)
```

```python
#lecture du resultat
df['age'].value_counts()
```

```python
#notre pourcentage
cut_tree = df['age'].value_counts()[0] 
age = df['age'].value_counts()[1] 
total_tree = cut_tree + age 
float (age) / total_tree 
```

```python
#on group
group_by_essence = df.groupby(['LIBELLEFRANCAIS','age'])
```

```python
group_by_essence.size()
```

```python
#etablir un tableau
count_tree_cut = group_by_essence.size().unstack()
count_tree_cut
```

```python
count_tree_cut = df.groupby(['LIBELLEFRANCAIS','age']).size().unstack().reset_index()
count_tree_cut[:5]
```

```python
count_tree_cut = count_tree_cut.fillna(0) 
count_tree_cut
```

```python
count_tree_cut['Nombre total'] = (count_tree_cut[False] + count_tree_cut[True])
count_tree_cut[:5]
```

```python
#tableau final avec false/true/nombre total/et proportion
count_tree_cut['Proportion'] = count_tree_cut[True] / count_tree_cut['Nombre total']
count_tree_cut[:4]
```

```python
#different plot
count_tree_cut['Proportion'].plot(legend=True,figsize = (100,40))
```
